# Checking Top 10 countries with respect to population of 2023

## ANALYSING INDIA'S POPULATION(1960-2024)

```python
india = df[df['Country Name'] == 'India']

india_details = india.melt(
    id_vars=['Country Name'], 
    value_vars=[str(y) for y in range(1960, 2025)],
    var_name='Year', 
    value_name='Population'
)
india_details['Year'] = india_details['Year'].astype(int)
india_details['Population'] = pd.to_numeric(india_details['Population'], errors='coerce')
```

## Bar Chart: India's Total Population Over Time

```python
plt.figure(figsize=(16, 9))
sns.barplot(x='Year', y='Population', data=india_details, color='green')
plt.title("India's Total Population (1960–2024)")
plt.xlabel("Year",size = 23)
plt.ylabel("Population",size = 23)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Regional & Global Population Trends

```python
regions = [
    'East Asia & Pacific', 'Europe & Central Asia', 'South Asia',
    'Sub-Saharan Africa', 'Latin America & Caribbean', 'North America',
    'Middle East & North Africa'
]

region_df = df[df['Country Name'].isin(regions)]

region_long = region_df.melt(
    id_vars=['Country Name'], 
    value_vars=[str(y) for y in range(2000, 2025)],
    var_name='Year',
    value_name='Population'
)
region_long['Year'] = region_long['Year'].astype(int)
region_long['Population'] = pd.to_numeric(region_long['Population'], errors='coerce')
```

## Line Chart: Population Over Time by Region (2000–2024)

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x='Year', y='Population', hue='Country Name', data=region_long, marker='o')
plt.title("Population Growth by World Bank Region (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
